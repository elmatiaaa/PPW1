```python
import pandas as pd
import requests
import csv
from bs4 import BeautifulSoup
```

# **Crawling Berita**

Crawling berita kali ini terdapat 3 kategori:
1. Otomotif
2. Megapolitan
3. Tren yang terjadi

Cara crawling yang saya lakukan disini yaitu dengan mengambil judul dan isi beritanya saja. Untuk mengambil judul dan isi berita disini dapat diproses melalui class yang ada pada elemen web berita online.

Setiap kategori memiliki halaman berita yang berbeda-beda, dapat dimasukkan pada variabel n yang sudah saya sediakan.

## Otomotif

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

linkberita = 'https://otomotif.kompas.com/'
listpolitik = []
n = 41

for halaman in range(1, n+1):
  req = requests.get(linkberita)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('div', 'article__list clearfix') #class semua berita
  for it in items:
    link = it.find('a','article__link')['href'] #class perberita

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judul = soup2.find('h1', 'read__title').text    #class judul berita
    isi = "".join(soup2.find('div', 'read__content').text.split('\n'))

    listpolitik.append([judul,isi])
```

# Megapolitan

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

linkberita = 'https://megapolitan.kompas.com/'
listt = []
n = 50

for halaman in range(1, n+1):
  req = requests.get(linkberita)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('div', 'article__list clearfix') #class semua berita
  for it in items:
    link = it.find('a','article__link')['href'] #class perberita

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judul = soup2.find('h1', 'read__title').text    #class judul berita
    isi = "".join(soup2.find('div', 'read__content').text.split('\n'))

    listt.append([judul,isi])
```

Menggabungkan 2 data crawling berita politik dan nasional menjadi 1 data frame dan memberikan label.

```python
politik = pd.DataFrame(listpolitik, columns=['Judul', 'Isi'])
politik['Label'] = 'otomotif'

nasional = pd.DataFrame(listt, columns=['Judul', 'Isi'])
nasional['Label'] = 'megapolitan'

gabung = [politik, nasional]
combined_df = pd.concat(gabung, ignore_index=True)
combined_df
```

```python
combined_df.to_csv('otomega.csv',  index=False)
```

## Tren

```python
header={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

urlberita = 'https://www.kompas.com/tren?source=navbar'
beritatren = []
n = 50

for halaman in range(1, n+1):
  req1 = requests.get(urlberita)
  soup1 = BeautifulSoup(req1.text, 'html.parser')
  items = soup1.findAll('div', 'trenLatest__item clearfix') #class berita
  for it in items:
    link = it.find('a','tren__link')['href'] #class perberita (pada bagian judul)

    req2 = requests.get(link)
    soup2 = BeautifulSoup(req2.text, 'html.parser')
    judulF = soup2.find('h1', 'read__title').text    #class judul berita
    isiF = "".join(soup2.find('div', 'read__content').text.split('\n'))

    beritatren.append([judulF,isiF])
```

```python
tren = pd.DataFrame(beritatren, columns=['Judul', 'Isi'])
tren['Label'] = 'tren'

gabungsemua = [combined_df, tren]
keseluruhan = pd.concat(gabungsemua, ignore_index=True)
keseluruhan
```

## hasil ketiga Kategori

```python
keseluruhan.to_csv('3kategori.csv',  index=False)
```
